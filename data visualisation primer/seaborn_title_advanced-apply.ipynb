```python
reviews = pd.read_csv('winemag-data_first150k.csv' , index_col=0)
reviews.head()
```

```python
reviews.points.value_counts().sort_index().plot.bar(figsize=(12,6))
```

```python
reviews.points.value_counts().sort_index().plot.bar(figsize=(12,6) , 
                                                    color='mediumvioletred')
```

```python
reviews.points.value_counts().sort_index().plot.bar(figsize=(12,6) , 
                                                    color='mediumvioletred',
                                                    fontsize = 16)
```

```python
reviews.points.value_counts().sort_index().plot.bar(figsize=(12,6) , 
                                                    color='mediumvioletred',
                                                    fontsize = 16,
                                                    title='ranking title by person')
```

```python
#pandas不能直接设置标题的特性
import matplotlib.pyplot as plt
```

```python
ax = reviews.points.value_counts().sort_index().plot.bar(figsize=(12,6) , 
                                                    color='mediumvioletred',
                                                    fontsize = 16)

ax.set_title('set title prope' , fontsize=20)
```

```python
import seaborn as sns #使用seaborn关闭边界线
ax = reviews.points.value_counts().sort_index().plot.bar(figsize=(12,6) , 
                                                    color='mediumvioletred',
                                                    fontsize = 16)

ax.set_title('set title prope' , fontsize=20)
sns.despine(bottom=True , left=True)
```
