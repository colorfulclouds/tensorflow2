```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
reviews = pd.read_csv('winemag-data_first150k.csv' , 
                      index_col = 0)
```

```python
fig , axarr = plt.subplots(2,1 , figsize=(8,8))

sns.countplot(reviews.points , ax=axarr[0])

reviews.points.value_counts().sort_index().plot.bar(ax=axarr[1])
```

```python
#seaborn没有处理线或者线下区域的函数
#但是有一个kdeplot

#kde kernel density extimate
#统计上的平滑噪音方法
sns.kdeplot(reviews[reviews.price<200]['price'])
```

```python
reviews[reviews.price<200]['price'].value_counts().sort_index().plot.line()
```

```python
sns.kdeplot(reviews[reviews.price<200].loc[: , ['price' , 'points']].dropna().sample(5000))

#双变量图 耗时
```

```python
#distplot函数等于pandas中的hist
sns.distplot(reviews.points , bins= 10 , kde=False)
```

```python
#双变量函数
sns.jointplot(x='price' , y='points' , data=reviews[reviews.price<100])
```

```python
#只需要设置参数kind='hex'即可
sns.jointplot(x='price',y='points',data=reviews[reviews.price<100],
              kind='hex',gridsize=20)
```

```python
#箱线图
df = reviews[reviews.variety.isin(reviews.variety.value_counts().head(5).index)]

sns.boxplot(x='variety',y='points',data=df)
```

```python
sns.violinplot(x='variety' , y='points',
               data=df)

#小提琴图 一个小提琴的外边沿就是分布曲线 类似kde曲线
```
