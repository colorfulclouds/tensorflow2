```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
```

```python
reviews = pd.read_csv('winemag-data_first150k.csv')
reviews.head(3)
```

## subplotting

```python
#subplots函数可以绘制多个图

fig , axarr = plt.subplots(2 , 1 , figsize=(12,8))
```

```python
axarr
```

```python
#可以告诉pandas将图像绘制到哪一个ax 轴上
fig , axarr = plt.subplots(2 , 1 , figsize=(12,8))

reviews.points.value_counts().sort_index().plot.bar(ax=axarr[0])

reviews.province.value_counts().head(20).plot.bar(ax=axarr[1])
```

```python
#上述的表图是2*1
#可以任意多
fig , axarr = plt.subplots(2 , 2 , figsize=(12,  8))

reviews.points.value_counts().sort_index().plot.bar(ax=axarr[0][0])

reviews.province.value_counts().head(20).plot.bar(ax=axarr[1][1])

#变为2*2后 axarr变为list in list
```

```python
axarr
```

```python
fig , axarr = plt.subplots(2 , 2 , figsize = (12 , 8))

reviews.points.value_counts().sort_index().plot.bar(ax=axarr[0][0] , fontsize = 12,
                                                    color = 'mediumvioletred')
axarr[0][0].set_title('wine scores' , fontsize=18)


reviews.variety.value_counts().head(20).plot.bar(ax=axarr[1][0] , fontsize=12,
                                                 color='mediumvioletred')
axarr[1][0].set_title('wine varieties' , fontsize=18)


reviews.province.value_counts().head(20).plot.bar(ax=axarr[1][1] , fontsize=12,
                                                  color='yellow')
axarr[1][1].set_title('wine province' , fontsize=18)


reviews.price.value_counts().plot.hist(ax=axarr[0][1] , fontsize=12, 
                                       color='red')
axarr[0][1].set_title('wine price' , fontsize=18)



plt.subplots_adjust(hspace=0.3)

sns.despine()
```
