```python
import pandas as pd

%matplotlib inline
```

```python
reviews = pd.read_csv('winemag-data_first150k.csv' , index_col=0)
```

```python
reviews.head()
```

### 双变量绘图

```python
#scatter 散点图
reviews[reviews.price < 100].sample(100).plot.scatter(x='price' , y='points')
```

```python
reviews[reviews.price<100].plot.scatter(x='price' , y='points')
#如果不采样 就毫无pattern
```

```python
#hex plot
reviews[reviews.price<100].plot.hexbin(x='price' , y='points' , gridsize=15)
#从下图可以看出大部分的葡萄酒87.5分 价格为20$
```

```python
pokemon = pd.read_csv('pokemon.csv' , index_col=0)
pokemon.head()
```

```python
pokemon.plot.scatter(x='attack' , y='defense')
```

```python
pokemon.plot.hexbin(x='attack',y='defense' , gridsize=16)
```

```python
temp = pokemon.groupby(['generation']).mean()[['attack' , 'defense']]
temp.plot.bar(stacked=True)
```

```python
temp_ = pokemon.groupby('generation').mean()[['hp','attack','defense','sp_attack','sp_defense','speed']]
temp_.plot.line()
```
