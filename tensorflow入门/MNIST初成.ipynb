```python
import numpy as np
import tensorflow as tf
import input_data

from scipy.misc import imread
from scipy.misc import imsave
```

```python
mnist = input_data.read_data_sets('MNIST_data/' , one_hot = True)
```

```python
test_x , test_y = mnist.train.next_batch(1)
print(test_x.shape)
print(test_y.shape)

test_x_img = np.reshape(test_x , [28 , 28 ,1])
```

```python
W = tf.Variable(tf.zeros([784 , 10]))
b = tf.Variable(tf.zeros([1 , 10]))

x = tf.placeholder(tf.float32 , shape=[None , 784])
y = tf.placeholder(tf.float32 , shape=[None , 10])
```

```python
predict = tf.nn.softmax(tf.matmul(x , W) + b)

cross_entropy = -tf.reduce_sum(y*tf.log(predict))

train_step = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for i in range(1000):
    batch_xs, batch_ys = mnist.train.next_batch(100)

    sess.run(train_step, feed_dict={x: batch_xs, y: batch_ys})
    #print(sess.run(cross_entropy , feed_dict={x: batch_xs, y: batch_ys}))
```

```python
right_prediction = tf.equal(tf.argmax(y , 1) , tf.argmax(predict , 1))

p_number = tf.argmax(y , 1)
real_number = tf.argmax(predict , 1)

accuary = tf.reduce_mean(tf.cast(right_prediction , tf.float32))

#sess.run(accuary , feed_dict={x: mnist.test.images, y: mnist.test.labels})

l = sess.run([p_number , real_number] , feed_dict={x: mnist.test.images, y: mnist.test.labels})
```
