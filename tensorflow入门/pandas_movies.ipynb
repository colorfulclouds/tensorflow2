```python
#创建dataframe
#直接扔进去一个dict，dict的键将作为列名，对应键的值将作为列值。
#按列创建
temp_dict = {'score':[8.9 , 8.2 , 9.3],
            'category':['xuanyi' , 'dongzuo' , 'aiqing']}

temp_pd = pd.DataFrame(temp_dict)
temp_pd
```

```python
#按行创建
'''
按行创建一般需要指定列名，
使用 columns = ['score', 'category'] 来分别对列重命名，
否则默认的列名是以数字命 0、1 命名的
'''
row1 = [8.9 , 'xuanyi']
row2 = [8.2 , 'dongzuo']
row3 = [9.3 , 'aiqing']

temp_pd = pd.DataFrame([row1 , row2 , row3] , columns=['score' , 'category'])
temp_pd
```

```python
#上面两种做法 都可以取出title列的信息 但是返回类型不一样
'''
带有一个中括号的类型是 Series，带有两个中括号的类型是 DataFrame。
 DataFrame 类似于 Excel 中的表，
那么 Series 就是 Excel 表中的某一行或某一列的数据类型,
多个有相同索引的 Series 就可以组成一个 DataFrame。
'''
'''
所以，movie_pd['title'] 返回的是 name 属性值为 title 的 Series，
    和 DataFrame 没有任何联系了；而 movie_pd[['title']] 返回的是列名为 title 的 DataFrame。

当需要单独对某一行或某一列进行运算时，才会使用一对中括号的方式
'''
```

```python
#逻辑筛选 与或非
print(movie_pd[movie_pd['category'] == '剧情'][['title','score']])
```

```python
print(movie_pd[(movie_pd['rank']<=5) & (movie_pd['score']>9.0)][['title', 'score' , 'rank']])
```

```python
print(movie_pd[(movie_pd['release_date']>'2010-01-01') | (movie_pd['vote_count']>500000)])
```

```python
print(movie_pd[movie_pd['url'].isnull()])
```

```python
print(movie_pd[movie_pd['regions'].notnull()])
```

```python
#筛选某个字段的值在给定列表中
print(movie_pd[movie_pd['score'].isin([8.0 , 9.0 , 9.5])])
```
