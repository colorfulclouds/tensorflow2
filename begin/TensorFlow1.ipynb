```python
#=====================
#=====================
#=====================
#MNIST

import numpy as np
import tensorflow as tf
import input_data

mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
W = tf.Variable(tf.zeros([784 , 10]))
b = tf.Variable(tf.zeros([1,10]))

x = tf.placeholder("float" , [None , 784])
y = tf.placeholder("float" , [None , 10])
```

```python
y_hat = tf.nn.softmax(tf.matmul(x , W) + b)

cross_entropy = -tf.reduce_sum(y * tf.log(y_hat))

train_step = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for i in range(1000):
    batch_xs, batch_ys = mnist.train.next_batch(100)
    sess.run(train_step, feed_dict={x: batch_xs, y: batch_ys})
    print(sess.run(cross_entropy , feed_dict={x: batch_xs, y: batch_ys}))
```

```python
correct_prediction = tf.equal(tf.argmax(y_hat , 1) , tf.argmax(y , 1))
```

```python
arruracy = tf.reduce_mean(tf.cast(correct_prediction , "float"))
```

```python
print(arruracy)
```

```python
sess.run(arruracy , feed_dict={x:mnist.test.images , y:mnist.test.labels})
```
